```python
from os.path import split
def sum_all_nums():
  nums = input("Enter a list of numbers, separated by commas: ")
  numslist = nums.split(",")
  sumnums = sum([int(num) for num in numslist])
  print(sumnums)

sum_all_nums()
```

```python
(penguins_data
 .filter([0,10], axis = 0)
 ).head() #Filtrar filas y se usa axis=0
```

```python
(penguins_data
 .query('species == "Gentoo"')
) #Se usa query para consultar / pedirle a la base de datos algo en especÃ­fico
```

```python
(penguins_data
 .query('body_mass_g > 5000')
).head() #Igual se puede filtrar con .query() para hacer un anÃ¡lisis
```

```python
(penguins_data
 .query('bill_length_mm >= 45')
).head()
```

```python
(penguins_data
 .filter(['species', 'body_mass_g', 'sex'], axis = 1)
 .query('species == "Gentoo"')
 ).head(3) #Se pueden unir las dos prÃ¡cticas para filtrar aÃºn mÃ¡s la base de datos
```

```python
(penguins_data
 .query('species == "Gentoo" & body_mass_g < 5700')
 ).tail(7) #Ãšltimos 7 registros de los Gentoo con una masa corporal menor a 5700g
```

```python
(penguins_data
 .assign(LDRatio = lambda df: df['bill_length_mm'] / df['bill_depth_mm'])
 ) #Creamos funciÃ³n anÃ³nima para hacer nueva columna, usamos df como data frame para que sea asignado a penguins_data
```

```python
(penguins_data
 .assign(LDRatio = lambda df: df['bill_length_mm'] / df['bill_depth_mm'])
 .filter(['bill_length_mm', 'bill_depth_mm', 'LDRatio'], axis = 1)
 ).head()
```

```python
(penguins_data
 .sort_values('bill_length_mm', ascending = False)
 ).head(3) #Arreglamos los valores con .sort_values()
```

```python
(penguins_data
 .filter(['bill_length_mm', 'bill_depth_mm', 'body_mass_g'], axis = 1)
 .agg(['mean'])
 ) #.agg() muestra alguna variable estadÃ­stica para tu tabla
```

```python
mean_penguins_data = (penguins_data
 .filter(['bill_length_mm', 'bill_depth_mm', 'body_mass_g'], axis = 1)
 .agg(['mean'])
 )

mean_penguins_data #Lo podemos hacer una variable/funciÃ³n para guardarlo
```
